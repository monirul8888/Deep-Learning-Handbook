```python
from sklearn.datasets import make_classification
import numpy as np
```

```python
X, y= make_classification(n_samples=100,
                          n_features=2,
                          n_informative=1,
                          n_redundant=0,
                          n_classes=2,
                          n_clusters_per_class=1,
                          random_state=41,
                          hypercube=False,
                          class_sep=10)
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
plt.scatter(
    X[:, 0],      # First feature
    X[:, 1],      # Second feature
    c=y,          # Color points by class
    cmap="winter",
    s=100
)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python


def perceptron(X, y):

    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])

    lr = 0.1

    for i in range(1000):

        j = np.random.randint(0, 100)

        y_hat = 1 if np.dot(X[j], weights) >= 0 else 0

        weights = weights + lr * (y[j] - y_hat) * X[j]

    return weights[0], weights[1:]
```

```python
intercept_, cofec = perceptron(X,y)
```

```python
m = - (cofec[0]/cofec[1])
b = -intercept_/cofec[1]

x_input = np.linspace(-3, 3, 100)
y_input = m*x_input + b
```

```python
print(X[:,1].min(), X[:,1].max())
```

```python
plt.figure(figsize=(10,7))

# Decision Boundary
plt.plot(x_input,
         y_input,
         color="red",
         linewidth=3)

# Data Points
plt.scatter(X[:,0],
            X[:,1],
            c=y,
            cmap="winter",
            s=100)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
